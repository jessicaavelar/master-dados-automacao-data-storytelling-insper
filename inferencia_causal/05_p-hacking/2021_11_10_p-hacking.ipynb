Vamos utilizar os dados do artigo de Cesar Zucco (2013) para rodar um modelo tosco, como exemplo do que não fazer em uma pesquisa.

A base de dados se encontra no link x, e o dicionário dos dados pode ser encontrado [aqui](https://www.dropbox.com/s/anth1tqxwejqdr3/Replication%20Instructions.pdf?dl=0).

Queremos mostrar que o crescimento da população evangélica diminui a busca pelo bolsa família.

Primeiro tentaremos mostrar a relação entre os gastos do bolsa família e o crescimento da população envagélica.

Em seguida, acrescentaremos outras variáveis como controle do modelo para obter o resultado que estamos buscando.

```python
import os
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols
```

```python
pbf = pd.read_csv("pbf_completo.csv", sep = ",")
pbf.head()
```

```python
list(pbf)
```

```python
# Montando a variável de crescimento pentecostal utilizando os dados do censo de 2000 e 2010.

pbf["pent_delta"] = pbf["pent_2010"] - pbf["pent_2000"]
```

```python
# Criando o modelo de regressão simples, utilizando os gastos com bf dependente
# e crescimento da população evangélica (proporcional à população) como variável independente.

model_envang_1 = ols("spentH_2010_cct ~ pent_delta",
                     data = pbf)
m_declinio = model_envang_1.fit()
m_declinio.summary()
```

No modelo encontramos uma relação positiva, ou seja, o aumento da população evangálica e bolsa família possui uma correlação, o contrário do que quero mostrar.

p_valor = 0.020

Por uma convenção científica, p-valores maiores de 5% não são significantes e acima de 10% nem merecem menção. Ou seja, se tudo o que a gente precisasse era mostrar que não há uma relação entre PBF e pent_delta, nosso trabalho sujo já teria terminado!
 
Agora, adicionaremos uma variável de controle ao modelo.

Votação do Lula em 2002.

Podemos argumentar no nosso artigo que o governo petista busca ajudar com mais bolsa-familia lugares que votaram no Lula em 2002, e ao mesmo tempo uma votação expressiva do PT afugenta pentecostais. A gente pode sempre contar uma história.

```python
# Adicionando Lula_vs_2002, votação no primeiro turno de 2006

model_envang_2 = ols("spentH_2010_cct ~ pent_delta + lula_vs_2002",
                     data = pbf)
m_declinio = model_envang_2.fit()
m_declinio.summary()
```

Ainda é positiva a relação, mas encontramos um p_valor de 0.107.

Vamos utilizar a votação do Lula em 2006, trocando a variável, vamos entender o que acontece com nossa relação.

```python
# Adicionando lula_vs_2002, votação do Lula no primeiro turno de 2002.

model_envang_3 = ols("spentH_2010_cct ~ pent_delta + lula_vs_2006",
                    data = pbf)
m_declinio = model_envang_3.fit()
m_declinio.summary()
```

Agora encontramos uma relação negativa entre o crescimento da população evangélica e o recebimento de bolsa família. Ou seja, quanto mais cresce a população evangálica, menos pessoas necessitam do auxílio.

pent_delta = -0.0036

p_valor = 0.052

Aqui, estamos mais próximos de mostrar o que propusemos inicialmente, porém, não atingimos um p_valor menor que 5%.

Por isos, vamos utilizar a votação no segundo turno no Lular em 2006.

```python
model_envang_4 = ols("spentH_2010_cct ~ pent_delta + lula_vs_2006_2",
                    data = pbf)
m_declinio = model_envang_4.fit()
m_declinio.summary()
```

pent_delta = -0.0039

p_valor = 0.050


Em uma lista com 70 variáveis, selecionamos aquela que consegue provar o nosso ponto, estamos fazendo **p_hacking**. Fazendo pequenas mudanças no modelo, eu construo a relação que eu quero, utilizando, não necessariamente, coisas que realmente importam para o estudo ou para o problema de pesquisa.
